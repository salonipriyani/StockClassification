```python
for a in symbols_list:
    print(len(fundamental_data),a)
    url = "https://www.marketwatch.com/investing/stock/"+a+"/financials/balance-sheet"
    tables = pd.read_html(url)
    if tables:
        df1 = tables[0]
        df1 = df1.transpose()
        new_header = df1.iloc[0] #grab the first row for the header
        df1 = df1[1:] #take the data less the header row
        df1.columns = new_header
        if len(tables)>1:
            df2 = tables[1]
            df2 = df2.transpose()
            new_header = df2.iloc[0] #grab the first row for the header
            df2 = df2[1:] #take the data less the header row
            df2.columns = new_header
            if len(tables)>2:
                df3 = tables[2]
                df3 = df3.transpose()
                new_header = df3.iloc[0] #grab the first row for the header
                df3 = df3[1:] #take the data less the header row
                df3.columns = new_header
                result = pd.concat([df1, df2, df3], axis=1)
            else:
                result = pd.concat([df1, df2], axis=1)
        else:
            result = df1
        result.drop(result.tail(1).index,inplace=True) # drop last row
        for i in range(len(result)):
            for j in range(len(result.iloc[0])):
                if '(' in result.iloc[i,j] and ')' in result.iloc[i,j]:
                    result.iloc[i,j] = result.iloc[i,j].replace('(', '')
                    result.iloc[i,j] = result.iloc[i,j].replace(')', '')
                if ',' in result.iloc[i,j]:
                    result.iloc[i,j] = result.iloc[i,j].replace(',', '')
                if 'B' in result.iloc[i,j]:
                    result.iloc[i,j] = result.iloc[i,j].replace('B', '') 
                    result.iloc[i,j] = float(result.iloc[i,j])
                    result.iloc[i,j] = result.iloc[i,j] * 1000000000
                elif 'T' in result.iloc[i,j]:
                    result.iloc[i,j] = result.iloc[i,j].replace('T', '') 
                    result.iloc[i,j] = float(result.iloc[i,j])
                    result.iloc[i,j] = result.iloc[i,j] * 1000000000000
                elif 'M' in result.iloc[i,j]:
                    result.iloc[i,j] = result.iloc[i,j].replace('M', '') 
                    result.iloc[i,j] = float(result.iloc[i,j])
                    result.iloc[i,j] = result.iloc[i,j] * 1000000
                elif '%' in result.iloc[i,j]:
                    result.iloc[i,j] = result.iloc[i,j].replace('%', '') 
                    result.iloc[i,j] = float(result.iloc[i,j])
                    result.iloc[i,j] = result.iloc[i,j]/100
                elif '-' in result.iloc[i,j]:
                    result.iloc[i,j] = result.iloc[i,j].replace('-', '0') 
                    result.iloc[i,j] = float(result.iloc[i,j])   
                else:
                    result.iloc[i,j] = float(result.iloc[i,j])
        fundamental_data[a] = result
```
